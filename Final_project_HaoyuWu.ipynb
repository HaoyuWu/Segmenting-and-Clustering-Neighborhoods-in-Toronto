# Battle of the neighborhoods
### - Final project of Haoyu Wu

## Introduction

In this project, we would like to compare the neighborhoods of two cites: New York and Toronto. Specifically, we will examine whether it is a good idea to move to Toronto if you are originally live in New York for a similar neighborhood.

Suppose a person lived in Manhattan and enjoy the neighborhood there, mainly because of all the great amenities and venues, such as restaurant, pharmacies, parks, schools and so on. However, a job-offer with a better payment and greater career plan arrived and required the relocation to Toronto. Wouldn't it be great if it is able to determine neighborhoods on the city of Toronto that are exactly the same as the current neighborhood, and if not, perhaps similar neighborhoods that are at least closer to the new job? 

Therefore, in this capstone project I will use location data to answer the question by comparing the neighborhoods in terms of a service and searching for potential explanation of why a neighborhood is popular. Furthermore, I will segment the cities into different clusters and then using a combination of location data and machine learning to find similar neighborhoods in a big city like New York or Toronto.

## Data

There are three sources of data used in this project. First, brought and neighborhood data of both Toronto and New York (NYC). For NYC, we can access the data from New York University. But for Toronto, we need to scrape data from Wikipedia webpage and transform that into dataframe that are usable in Python. Second, coordinates data for each borough for both cities. It is extracted from Google geocoding API serve according to the combination of the keyword in “postal code” and “boroughs”. Third, location data for each neighborhood, which is scraped from Foursquare API. 

## Methodology

To compare the similarities of two cities, we decided to explore neighborhoods, segment them, and group them into clusters to find similar neighborhoods in a big city like New York and Toronto. To be able to do that, we need to cluster data which is a form of unsupervised machine learning: k-means clustering algorithm. 

k-means clustering is a method of vector quantization, that is popular for cluster analysis in data mining. The method aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean, serving as a prototype of the cluster. This results in a partitioning of the data space into Voronoi cells.

we will firstly explore neighborhoods in both cities on Foursquare API and get the most common venue categories in each neighborhood. Next, k-means clustering algorithm will be adopted to group the neighborhoods into clusters. Finally, we will visualize the neighborhoods in two cities and their emerging clusters, then compare the similarities and dissimilarities of two locations. 

## Result

The clusters results of two cities are distributed on Table 1 and Table 2. The visualized results are shown on Figure 1 and Figure 2. For Manhattan, 25 neighborhoods are accounted in cluster 0, 7 are accounted in cluster 1, 8 are accounted into cluster 4, the rest 2 are separately accounted in the cluster 2 and 3. Moreover, from figure 1 we also can see that the neighborhoods are dispersively distributed on the map, even if they are in the same clusters. That is to say, the neiborhoods are quite diversified in Manhattan, that you can usually find more than one place that have the similar fetures, which are located in different part of the city. 

However, the situation is quite different in Toronto. The neighborhoods are more similar than in Manhattan. There are 33 out of 37 neighbors are agglomerated in cluster 0. And the rest 4 clusters only contain one neighborhood respectively. In other words, those 33 neighborhoods have highly similar venues around them, but the rest 5 neighborhoods surrounded by less likely venues than others. Thus, if this person live in the clusters contain 32 regions are highly recommended to move within this clusters. Moreover, he/she may not satisfies the neighborhood if the new jobs are in located in the neighborhood of St.James Town, University of Toronto, Roselawn, Toronto Dominion Centre or Davisville.

Nest, we would like to compare the neighborhoods in Manhattan and Toronto. Hence, we combine the location data of two cities tighter and check which neighborhoods would be put into a same cluster category. The results are listed in Table 3. Cluster 1 contains some US neighbors, but all Toronto neighbors. That means, if put two cities under consideration, all neighborhoods in Toronto are put into a similar category, which could be replicate by the Manhattan neighbors in the same cluster. That is to say if moving from Toronto to Manhattan, no matter which neighrhood you are originally from, you can find a similar neighborhood in Manhattan. But if you live in Manhattan, one the person from the cluster 3 would find a neighborhoods that fit for him/her. 

## Conclusion

In this project, through a k-means cluster algorithm we separate the neighborhood into 5 clusters, which have similar neighborhoods around them. We discover that if a person would like to move within the Manhattan, we cam always find a perfect substitution unless he/she lives in Little Italy or Carnegie Hill. But for Toronto, most neighborhoods own analogous characteristics that relocation is easier to be accomplished and adapted. If moving internationally, a Canadian could find similar neighborhood in Manhattan most of time, On the other hand, a New Yorker could always reconcile himself/herself to the life in Toronto, since every neighborhood in Manhattans are clustered in a category with at least one neighborhood in Toronto. We can see from this case that New York, as the most diversified city of the world, contains almost any type of live style and different preferences. And due to the same perspective, the person from New York can adapt to new environment more easily than people from other city, that they have experiences the life around the world just within New York.
